## Problem Statement:

* Lung cancer can cause complications, such as: Shortness of breath. People with lung cancer can experience shortness of breath if cancer grows to block the major airways. Lung cancer can also cause fluid to accumulate around the lungs, making it harder for the affected lung to expand fully when you inhale.The effectiveness of cancer prediction system helps the people to know their cancer risk with low cost and it also helps the people to take the appropriate decision based on their cancer risk status. ## Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1 = data.copy(deep=True)
# convert categorical columns to numerical values
#{'LUNG_CANCER':{'No':0,'Yes':1},'GENDER':{'Male':1,'Female':0}}
```

```python
GENDER = le.fit_transform(data['GENDER'])
LUNG_CANCER = le.fit_transform(data['LUNG_CANCER'])
```

```python
data['GENDER']= GENDER
data['LUNG_CANCER']= LUNG_CANCER
```

```python
data.head()
```

## Train Test Split:

```python
# separating the data and label
X = data.drop(columns=['LUNG_CANCER'],axis=1)
y = data['LUNG_CANCER']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## SVM Algorithm

```python
# Fitting classifier to the Training set
from sklearn.svm import SVC
classifier = SVC(random_state=0) # for non-linear model use this parametre kernel='rbf'
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 'C':[0.1,1,100,1000],'kernel':['rbf','poly','sigmoid','linear'],'degree':[1,2,3,4,5,6]}
```

```python
grid= GridSearchCV(SVC(),param_grid, verbose=3, n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
print(grid.best_params_)
```

```python
svc =SVC(C=1000, gamma=1,degree = 5,kernel = 'linear',probability=True)
svc.fit(X_train,y_train)
```

```python
#ac_svc = accuracy_score(y_test,y_pred)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(svc, X, y, cv =5)
ac_svc = np.mean(scores)
```

```python
print('Model accuracy: ',ac_svc)
```

```python
# Confusion matrix for SVM
from sklearn.metrics import confusion_matrix

y_pred_svc = svc.predict(X_test)
cf_matrix = confusion_matrix(y_test, y_pred_svc)
sns.heatmap(cf_matrix, annot=True)
plt.title("Confusion Matrix for SVM", fontsize=14, fontname="DejaVu Sans", y=1.03)
```
